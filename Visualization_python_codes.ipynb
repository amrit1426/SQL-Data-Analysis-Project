```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the data
top_paying_jobs = pd.read_csv('1_top_paying_jobs_op_data.csv')
top_paying_jobs
```

```python
# Plotting the SQL query output

df_plot_data1 = top_paying_jobs[['job_title', 'salary_year_avg']].set_index('job_title')
df_plot_data1.index = df_plot_data1.index.str.slice(0, 20) + "..."

# Plotting bar chart
figure = plt.figure(figsize=[15, 7])
df_plot_data1.plot.bar(color='#1B9E77')

# Aesthetics
plt.xlabel('Job Titles', size=14)
plt.ylabel('Average Yearly Salary', size=14)
plt.title('Top 10 Paying Jobs', size=16)
plt.xticks(rotation=45, ha='right')
plt.legend().remove()
plt.tight_layout()

# Save as JPEG
plt.savefig('Q1.jpeg')

# Display output
plt.show()
```

```python
# Loading the top paying skills data
top_paying_skills = pd.read_csv('2_top_paying_job_skills_op_data.csv')
# Plotting the top paying skills data
top_10_skills = top_paying_skills[['skills', 'salary_year_avg']].set_index('skills')

# Plotting bar chart
figure = plt.figure(figsize=[15, 7])
top_10_skills.head(10).plot.bar(color='#1B9E77')

# Aesthetics
plt.xlabel('Skills', size=14)
plt.ylabel('Average Yearly Salary', size=14)
plt.title('Top 10 Paying Skills', size=16)
plt.xticks(rotation=45, ha='right')
plt.legend().remove()
plt.tight_layout()

# Save as JPEG
plt.savefig('Q2.jpeg')

# Display output
plt.show()
```

```python
# Loading the top paying skills data
top5_demanding_skills = pd.read_csv('3_top_demanding_skills_op_data.csv')
# Plotting the top paying skills data
top_5_skills = top5_demanding_skills[['skills', 'skills_demand_count']].set_index('skills')

# Plotting bar chart
figure = plt.figure(figsize=[15, 7])
top_5_skills.plot.bar(color='Green')

# Aesthetics
plt.xlabel('Skills', size=14)
plt.ylabel('Job Count', size=14)
plt.title('Top 5 Most Demanding Skills', size=16)
plt.xticks(rotation=90, ha='center')
plt.legend().remove()
plt.tight_layout()

# Save as JPEG
plt.savefig('Q3.jpeg')

# Display output
plt.show()
```

```python
# Loading the top paying skills data
df_plot = top_skills_vs_salary = pd.read_csv('4_top_skills_based_on_salary_op_data.csv').head(15)

# Create figure
plt.figure(figsize=(14, 7))

# Horizontal bar plot
plt.bar(
    df_plot['skills'],
    df_plot['avg_salary']
)

# Labels and title
plt.ylabel('Average Salary', size=14)
plt.xlabel('Skills', size=14)
plt.title('Average Salary by Skill', size=16)
plt.xticks(rotation=90, ha='center')

# Improve layout
plt.tight_layout()

# Save image
plt.savefig('Q4.jpeg', dpi=300)

# Show plot
plt.show()
```

```python
most__optimal_skills = pd.read_csv('5_most__optimal_skills_op_data.csv').head() #.set_index('skills')
demand_color = '#4E79A7'
salary_color = '#F28E2B'

df_plot = most__optimal_skills.sort_values(by='demand_count', ascending=False)

fig, my_canvas = plt.subplots(figsize=(14, 5), nrows=2, sharex=True)

# Demand Plot
my_canvas[0].bar(
    df_plot['skills'],
    df_plot['demand_count'],
    color=demand_color
)
my_canvas[0].set_ylabel('Demand Count', size=12)
my_canvas[0].set_title('Skill Demand vs Average Salary', size=14)

# Salary Plot
my_canvas[1].bar(
    df_plot['skills'],
    df_plot['avg_salary'],
    color=salary_color
)
my_canvas[1].set_ylabel('Average Salary', size=12)
my_canvas[1].set_xlabel('Skills', size=12)

plt.xticks(rotation=0, ha='center')
plt.tight_layout()
plt.savefig('Q5.jpeg', dpi=300)
plt.show()
```
